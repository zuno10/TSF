```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import requests
```

```python
url = 'http://bit.ly/w-data'
df = pd.read_csv(url)
```

```python
## Preparing Data
X = df.iloc[:,0].values
y = df.iloc[:,1].values
X.shape,y.shape
```

```python
sns.scatterplot(x=X,y=y)
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Hours and Scores')
```

```python
# ## splitting the into train, test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
sns.scatterplot(x=X_train,y=y_train)
sns.scatterplot(x=X_test,y=y_test)
plt.legend(['Training Data','Testing Data'])
```

```python
## linear Regression
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# X = np.array(X).reshape(1,-1)
# y = np.array(y).reshape(1,-1)
```

```python
## training Regression model
model.fit(X_train.reshape(-1,1),y_train)
```

```python
sns.scatterplot(x=X,y=y)
plt.plot(X, model.coef_*X + model.intercept_, color='green')
```

## Prediction & Evaluation

```python
y_pred = model.predict(X_test.reshape(-1,1))
pd.DataFrame({'Actual' : y_test, 'Prediction' :y_pred})
```

```python
from sklearn import metrics
print(f'Mean Absolute Error : {metrics.mean_absolute_error(y_pred,y_test)}')
```
